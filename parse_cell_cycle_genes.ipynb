```python
import pandas as pd
```

```python
current_phase = None
gene_list = []

with open("cell_cycle_mouse.tsv") as fh:
    for line in fh:
        line = line.strip()
        if line.startswith('#'):
            current_phase = line.split('#')[1]
        else:
            genes = line.replace('"', '').split(' ')
            for gene in genes:
                if gene != '':
                    gene_list.append([gene, current_phase])

print(gene_list)
```

```python
df = pd.DataFrame(gene_list, columns=["ensembl_code", "phase"])
```

```python
mapper = pd.read_csv('mouse_annotation_map.tsv', sep='\t')
```

```python
final = pd.merge(df, mapper, how='inner', left_on='ensembl_code', right_on='Gene ID')
```

```python
final.to_csv('mouse_cell_cycle.tsv', sep='\t', index=False)
```
